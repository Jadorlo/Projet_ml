## Exploration des données

L'exploration des données permet de mieux appréhender le sujet que l'on va traiter. Cette analyse est réalisée avec des graphiques matplotlib, la construction de variables nouvelles pour essaayer de capter certains mouvements comme les variations en pourcentage par jour et mois, consigné dans des tableux.

```python
def exploration_donnees(df : pd.DataFrame):
    
    print("Analyse des variables numériques :")
    display(df.describe().round(2))
    
    print("Analyse des variables temporelles :")
    df1 = df.copy()
    df1["date"] = pd.to_datetime(df1["date_utc"], format="%Y-%m-%d %H:%M:%S+00:00")
    display(df1["date"].describe())
    
    # Calcul de l'intervalle de temps
    intervalle = max(df1['date']) - min(df1['date'])

    # Conversion en années, mois et jours
    annees = intervalle.days // 365
    mois = (intervalle.days % 365) // 30
    jours = (intervalle.days % 365) % 30

    print(f"Intervalle de temps des données : {annees} années, {mois} mois, et {jours} jours")
    
    print("\nAnalyse graphique :")
    print("Visualisation de la tendance du Bitcoin :")
    df1.plot(x="date_utc", y = ["Close_price"], kind = 'line', figsize=(20,7))
    plt.show()
    
    print("\nAnalyse avec Bougie Japonaise journalières :")
    print("Les bougies japonaises permettent de voir")

    # Tri des données par date (important pour l'ordre chronologique)
    df1 = df1.sort_values(by="date_utc")
    df1["date_utc"] = df1["date_utc"].str[0:10]

    # Taille de la figure
    plt.figure(figsize=(25, 8))

    # Largeurs des corps des bougies et des mèches
    width = 0.65  # Largeur des corps des bougies
    width2 = 0.35  # Largeur des mèches (High et Low)

    # Couleurs pour les bougies
    col1 = 'green'  # Couleur pour les hausses
    col2 = 'red'    # Couleur pour les baisses

    # Parcourir chaque ligne pour dessiner les bougies dans l'ordre des dates
    for index, row in df1.iterrows():
        color = col1 if row["Close_price"] >= row["Open_price"] else col2

        # Tracer le corps de la bougie
        plt.bar(row["date_utc"], row["Close_price"] - row["Open_price"], width, 
                bottom=min(row["Close_price"], row["Open_price"]), color=color, edgecolor=color)
        
        # Tracer les mèches (High et Low)
        plt.vlines(row["date_utc"], row["Low_price"], row["High_price"], color=color, linewidth=width2)

    # Ajouter un titre et des labels
    plt.title("Graphique en Bougie Japonaise", fontsize=16)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Prix", fontsize=12)

    # Rotation des ticks sur l'axe des X pour une meilleure lisibilité
    plt.xticks(rotation=45, ha='right')

    # Ajustement de l'affichage des ticks sur l'axe des X
    plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))

    # Afficher le graphique
    plt.tight_layout()
    plt.show()
    
    print("\nÉvolution par mois :")
    df2 = df.copy()
    df2["date_utc"] = df2["date_utc"].str[0:7]
    df2["evolution_prct"] = df1["Close_price"].pct_change()*100
    df2 = df2.groupby(by = ["date_utc"])["evolution_prct"].mean()
    df2 = pd.DataFrame(df2)
    display(df2)
    df2.plot(kind='line')

exploration_donnees(base)
```

## Prétraitements des données

```python
def traitement_donnees(df : pd.DataFrame):
    df1 = pd.DataFrame()
    df1 = df.copy()
    df1["date_utc"] = df1["date_utc"].astype(str).str[:10]
    
    df_jour_mois = []
        
    for i in df1["date_utc"]:
        date_string = i
        date_object = datetime.strptime(date_string, "%Y-%m-%d")
        jour_semaine = date_object.weekday()
        jour_nom = date_object.strftime("%A")
        mois_num = date_object.month  
        # print(f"La date {date_string} correspond à : {jour_nom} (jour {jour_semaine})")
        
        df_jour_mois.append({
            "date_utc" : date_string,
            "lundi": 1 if jour_semaine == 0 else 0,
            "mardi" : 1 if jour_semaine == 1 else 0 ,
            "mercredi" : 1 if jour_semaine == 2 else 0,
            "jeudi" : 1 if jour_semaine == 3 else 0,
            "vendredi" : 1 if jour_semaine == 4 else 0,
            "samedi" : 1 if jour_semaine == 5 else 0,
            "dimanche" : 1 if jour_semaine == 6 else 0,
            "janvier": 1 if mois_num == 1 else 0,
            "fevrier": 1 if mois_num == 2 else 0,
            "mars": 1 if mois_num == 3 else 0,
            "avril": 1 if mois_num == 4 else 0,
            "mai": 1 if mois_num == 5 else 0,
            "juin": 1 if mois_num == 6 else 0,
            "juillet": 1 if mois_num == 7 else 0,
            "aout": 1 if mois_num == 8 else 0,
            "septembre": 1 if mois_num == 9 else 0,
            "octobre": 1 if mois_num == 10 else 0,
            "novembre": 1 if mois_num == 11 else 0,
            "decembre": 1 if mois_num == 12 else 0
        })
    
    df_jour_mois = pd.DataFrame(df_jour_mois)
    df1["evolution_prct"] = df1["Close_price"].pct_change()*100
    df_resultat = pd.merge(df1, df_jour_mois, on=["date_utc"])
    
    return df_resultat
    
df_traite = traitement_donnees(base)
df_traite
```

### Tensorflow

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
```

```python
df_traite = pd.read_csv('..\data\dataset_v1.csv')

# Colonnes utilisées comme entrées (features)
feature_columns = [
    'Open_price', 'High_price', 'Low_price', 'Volume', 'Quote_volume',
    'Nb_trades', 'Taker_buy_base_volume', 'Taker_buy_quote_volume',
    'lundi', 'mardi', 'mercredi', 'jeudi', 'vendredi', 'samedi', 'dimanche',
    'janvier', 'fevrier', 'mars', 'avril', 'mai', 'juin', 'juillet', 'aout',
    'septembre', 'octobre', 'novembre', 'decembre'
]

# Préparer les données
X = df_traite[feature_columns].values
y = df_traite['Close_price'].values.reshape(-1, 1)

# Normalisation des données
scaler_features = MinMaxScaler()
scaler_target = MinMaxScaler()

X = scaler_features.fit_transform(X)
y = scaler_target.fit_transform(y)

# Créer des séquences pour LSTM
def create_sequences(X, y, sequence_length):
    sequences = []
    targets = []
    for i in range(len(X) - sequence_length):
        sequences.append(X[i:i + sequence_length])
        targets.append(y[i + sequence_length])
    return np.array(sequences), np.array(targets)

sequence_length = 10
X_seq, y_seq = create_sequences(X, y, sequence_length)

# Division chronologique en train/test
train_size = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]
```

```python
X_seq.shape, y_seq.shape
```

```python
# Définir le modèle
model = Sequential([
    LSTM(128, input_shape=(sequence_length, X_train.shape[2]), return_sequences=True),
    tensorflow.keras.layers.Dropout(0.2),
    Dense(64, activation='relu'),
    tensorflow.keras.layers.Dropout(0.2),
    Dense(32, activation='relu'),
    tensorflow.keras.layers.Dropout(0.2),
    Dense(16, activation='relu'),
    tensorflow.keras.layers.Dropout(0.2),
    Dense(1)  # Prédiction d'une seule valeur
])

model.summary()
cp = ModelCheckpoint('model2.keras', save_best_only = True)
# Compilation du modèle
model.compile(optimizer= Adam(learning_rate = 0.01), loss='mse', metrics = [RootMeanSquaredError()])
```

```python
# Entraînement
model.fit(
    X_train, y_train,
    validation_split=0.2,  # Utiliser 10% des données d'entraînement pour validation
    epochs=100,
    callbacks = [cp],
    batch_size=10,
    verbose=1)
```

```python
model = load_model('model2.keras')
# Évaluation sur l'ensemble de test
y_pred = model.predict(X_test)

# Dénormaliser les valeurs
y_pred_rescaled = scaler_target.inverse_transform(y_pred)
y_test_rescaled = scaler_target.inverse_transform(y_test)

# Visualisation des résultats
plt.figure(figsize=(12, 6))
plt.plot(y_test_rescaled, label="Vraies valeurs", color="blue")
plt.plot(y_pred_rescaled, label="Prédictions", color="red")
plt.legend()
plt.title("Prédictions LSTM avec TensorFlow")
plt.show()
```
